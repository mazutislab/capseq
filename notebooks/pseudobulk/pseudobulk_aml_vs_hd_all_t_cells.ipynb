```python
# Store raw counts in layers
pdata.layers["counts"] = pdata.X.copy()

# Normalize, scale and compute pca
sc.pp.normalize_total(pdata, target_sum = 1e4)
sc.pp.log1p(pdata)
sc.pp.scale(pdata, max_value = 10)
sc.tl.pca(pdata)

# Return raw counts to X
dc.swap_layer(pdata, "counts", X_layer_key=None, inplace=True)
```

```python
# Select T cells profiles
print(select_cell_type)
t_cells = pdata[pdata.obs["cell_type_simplified"] == select_cell_type].copy()
```

```python
dc.plot_filter_by_expr(t_cells, group = "sample", min_count = 5, min_total_count = 10)
```

```python
# Obtain genes that pass the thresholds
genes = dc.filter_by_expr(t_cells, group = "sample", min_count = 5, min_total_count = 10)

# Filter by these genes
t_cells = t_cells[:, genes].copy()
t_cells
```

```python
# Import DESeq2
from pydeseq2.dds import DeseqDataSet, DefaultInference
from pydeseq2.ds import DeseqStats
```

```python
t_cells.obs["sample"].unique()
```

```python
# Build DESeq2 object
inference = DefaultInference(n_cpus = 8)
dds = DeseqDataSet(
    adata = t_cells,
    design_factors = "sample",
    ref_level = ["sample", select_control_sample],
    refit_cooks = True,
    inference = inference,
)
```

```python
# Compute LFCs
dds.deseq2()
```

```python
# Extract contrast between AML vs HD
stat_res = DeseqStats(
    dds,
    contrast=["sample", select_condition_sample, select_control_sample],
    inference = inference,
)
```

```python
# Compute Wald test
stat_res.summary()
```

```python
# Extract results
results_df = stat_res.results_df
results_df
```

```python
# Export volcano genes
results_df.to_csv("dge/pseudobulk_volcano_t_cells_AML_vs_HD.csv")


results_df_sorted = results_df.copy()
results_df_sorted[results_df_sorted["padj"] < 0.05].sort_values(by = "log2FoldChange", ascending=False, key=abs).to_excel("dge/pseudobulk_volcano_t_cells_AML_vs_HD_filtered_by_padj.xlsx")

```
